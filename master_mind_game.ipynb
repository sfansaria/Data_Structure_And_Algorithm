```python
import random

num = random.randrange(1000, 10000)

n = int(input("Guess the 4 digit number: "))

if n == num:
  print("Hurrah....I guessed the number in the 1st try....I am the mastermind")
else:
  chances = 0
  while n != num:
    chances += 1
    count = 0
    n = str(n)
    num = str(num)
    correct = ['X']*4

    for i in range(0, 4):
        if (n[i] == num[i]):
          count += 1
          correct[i] = n[i]
        else:
          continue

    print("NOt quite the number. But you did get",count, "digit(s) correct!")


    print('\n')
    print('\n')
    n = int(input("Enter your next choice of numbers: "))

  if n == num:
    chances += 1
    print("You have become a mastermind!")
    print("It took you only",chances,"tries.")
```
